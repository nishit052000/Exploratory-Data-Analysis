Categorizing vote count column

```python
def categorise_col(df,cols, labels):
    
    edges=[df[cols].describe()['min'],
          df[cols].describe()['25%'],
          df[cols].describe()['50%'],
          df[cols].describe()['75%'],
          df[cols].describe()['max']]
    df[cols]=pd.cut(df[cols],edges,labels=labels,duplicates='drop')
    return df
```

```python
labels=["not_popular",'below_avg','average','popular']
categorise_col(df,'Vote_Average',labels)
df['Vote_Average'].unique()
```
